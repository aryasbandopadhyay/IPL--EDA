```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
data=pd.read_csv('Deliveries.csv')
data
```

```python
bat=data.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
ball=data.groupby('batsman')['ball'].count().reset_index()
```

```python
ball=ball[ball['ball']>200]
```

```python
batball=bat.merge(ball,on='batsman')
```

```python
batball
```

```python
batball['ball']=batball['batsman_runs']/batball['ball']*100
```

```python
batball
```

```python
y=data.groupby('batsman')['match_id'].nunique().reset_index()
```

```python
batball=batball.merge(y,on='batsman')
```

```python
batball['match_id_x']=batball['batsman_runs']/batball['match_id_y']
```

```python
batball
```

```python
X=batball.iloc[:,2].values
```

```python
y=batball.iloc[:,3].values
```

```python
plt.scatter(X,y)
```

```python
from sklearn.cluster import KMeans
```

```python
x=batball.iloc[:,2:4].values
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit(x)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,11),wcss)
```

```python
km=KMeans(n_clusters=3)
```

```python
ymeans=km.fit_predict(x)
ymeans
```

```python
plt.scatter(x[ymeans==0,0],x[ymeans==0,1],color='red')
plt.scatter(x[ymeans==1,0],x[ymeans==1,1],color='blue')
plt.scatter(x[ymeans==2,0],x[ymeans==2,1],color='yellow')
```
